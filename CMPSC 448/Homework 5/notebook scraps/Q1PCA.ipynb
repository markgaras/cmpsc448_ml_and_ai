```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import preprocessing
import numpy as np 
from sklearn.decomposition import PCA
```

```python
iris = load_iris() 
X = iris.data 
y = iris.target
x = np.array(X)
```

```python
pca1 = PCA(n_components=1) 
pca1.fit(x) 

print("\n n component = 1 \n")
print(" exp var ratio: ", pca1.explained_variance_ratio_) 
print(" exp var: ", pca1.explained_variance_)

pca2 = PCA(n_components=2) 
pca2.fit(x) 

print("\n n component = 2 \n")
print(" exp var ratio: ", pca2.explained_variance_ratio_) 
print(" exp var: ", pca2.explained_variance_)

pca3 = PCA(n_components=3) 
pca3.fit(x) 

print("\n n component = 3 \n")
print(" exp var ratio: ", pca3.explained_variance_ratio_) 
print(" exp var: ", pca3.explained_variance_)

pca4 = PCA(n_components=4) 
pca4.fit(x) 

print("\n n component = 4 \n")
print(" exp var ratio: ", pca4.explained_variance_ratio_) 
print(" exp var: ", pca4.explained_variance_)
```

```python
x_scaled = preprocessing.scale(x)
x_min_max = preprocessing.MinMaxScaler().fit_transform(x_scaled)
x_normalized = preprocessing.normalize(x_min_max, norm='l2')

print("\n after processing, n component = 1 \n")
pca11 = PCA(n_components=1) 
pca11.fit(x_normalized) 
print(" exp var ratio: ", pca11.explained_variance_ratio_) 
print(" exp var: ", pca11.explained_variance_)

print("\n after processing, n component = 2 \n")
pca12 = PCA(n_components=2) 
pca12.fit(x_normalized) 
print(" exp var ratio: ", pca12.explained_variance_ratio_) 
print(" exp var: ", pca12.explained_variance_)

print("\n after processing, n component = 3 \n")
pca13 = PCA(n_components=3) 
pca13.fit(x_normalized) 
print(" exp var ratio: ", pca13.explained_variance_ratio_) 
print(" exp var: ", pca13.explained_variance_)

print("\n after processing, n_components = 4 \n")
pca14 = PCA(n_components=4) 
pca14.fit(x_normalized) 
print(" exp var ratio: ", pca14.explained_variance_ratio_) 
print(" exp var: ", pca14.explained_variance_)
```

```python
pca12 = PCA(n_components=2) 
a = pca12.fit_transform(x_normalized) 
print(" exp var ratio: ", pca12.explained_variance_ratio_) 
print(" exp var: ", pca12.explained_variance_)

plt.scatter(a[:,0], a[:,1], c=y, edgecolor='k')
plt.show()
```

```python
kMeans = KMeans(n_clusters=3, init='k-means++')
k = kMeans.fit(a)
plt.scatter(a[:,0], a[:,1], c=k.labels_, edgecolors='k')
plt.show()
```
